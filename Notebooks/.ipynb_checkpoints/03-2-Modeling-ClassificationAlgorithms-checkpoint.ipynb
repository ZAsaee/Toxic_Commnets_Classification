## 2. Performance Evaluation Functions

## 2.1. ROC Curve

```python
def roc_plot(estimator, X, y, classifier_name=None, n_classes=2):
    
    categories = ['Non-toxic', 'Toxic']
    fig = plt.figure(figsize=(14,4))
    plt.rcParams.update({'font.size': 12})
    
    for i in range(n_classes):
        
        ax = fig.add_subplot(1, 2, i+1)
        
        display = RocCurveDisplay.from_estimator(estimator=estimator, X=X, y=y, pos_label=i, 
                                      response_method='predict_proba', ax=ax, name=classifier_name, 
                                                 color='r', lw=2)
        
        ax.plot([0,1], [0,1], color='navy', linestyle='--', lw=2)

        ax.set_xlim([0.0, 1.0])
        ax.set_ylim([0.0, 1.0])
        ax.set_xlabel('False Positive Rate')
        ax.set_ylabel('True Positive Rate')
        ax.set_title('Reciever Operating Characteristic (ROC) - {} '.format(categories[i]))
```

## 2.2. Precision/Recall Curve

```python
def prec_recall_plot(estimator, X, y, classifier_name=None, n_classes=2):
    
    categories = ['Non-toxic', 'Toxic']
    fig = plt.figure(figsize=(14,4))
    plt.rcParams.update({'font.size': 12})
    
    f_scores = np.linspace(0.2, 0.8, num=4)
    lines, labels = [], []
    
    for i in range(n_classes):
        
        ax = fig.add_subplot(1, 2, i+1)
        
        display = PrecisionRecallDisplay.from_estimator(estimator=estimator, X=X, y=y, pos_label=i, 
                                      response_method='predict_proba', ax=ax, name=classifier_name,
                                                       color='g')

        ax.set_xlim([0.0, 1.0])
        ax.set_ylim([0.0, 1.0])
        ax.set_xlabel('Recall')
        ax.set_ylabel('Precision')
        ax.set_title('Precision-Recall Curve - {} '.format(categories[i]))
```

```python
 def prec_recall_plot_multi(estimator, X_test, y_test, classifier_name=None, n_classes=2):
    
    color = ["teal", "deeppink"]
    categories = ['Non-toxic', 'Toxic']
    plt.rcParams.update({'font.size': 12})
    
    _, ax = plt.subplots(figsize=(8, 6))
    f_scores = np.linspace(0.2, 0.8, num=4)
    lines, labels = [], []
    for f_score in f_scores:
        x = np.linspace(0.01, 1)
        y = f_score * x / (2 * x - f_score)
        (l,) = plt.plot(x[y >= 0], y[y >= 0], color="gray", alpha=0.5)
        plt.annotate("f1={0:0.1f}".format(f_score), xy=(0.9, y[45] + 0.02))
        
    
    for i in range(n_classes):
        
        display = PrecisionRecallDisplay.from_estimator(estimator=estimator, X=X_test, y=y_test, pos_label=i, 
                                      response_method='predict_proba', ax=ax,
                                      name='Precision-recall for {} class'.format(categories[i]),
                                                       color=color[i])
    ax.set_title('Precision-Recall Curve')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.02])
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    
    handles, labels = display.ax_.get_legend_handles_labels()
    handles.extend([l])
    labels.extend(["iso-f1 curves"])
    ax.legend(handles=handles, labels=labels, loc="lower left")
    plt.show()
```

## 3. Classification and HyperParameter Tuning Function

```python
def bayessearchcv(pipeline, param_grid, scoring, cv, n_iter=50, n_jobs=1, 
                 classifier_name=None, fit_params=None, 
                  X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test):
    
    print("X train type and shape: ", type(X_train), X_train.shape)
    print("y train type and shape: ", type(y_train), y_train.shape)
    
    print("X test type and shape: ", type(X_test), X_test.shape)
    print("y test type and shape: ", type(y_test), y_test.shape)
    print('-----------------------------------')
    
    search = BayesSearchCV(estimator=pipeline, search_spaces=param_grid, n_iter=n_iter,
                          scoring=scoring, cv=cv, n_jobs=n_jobs, verbose=1, 
                           random_state=12, fit_params=fit_params)
    
    print("Performing the bayesian grid search:")
    print("pipeline:", [name for name, _ in pipeline.steps])
    print("parameters:")
    pprint(param_grid)
    print("------------------------------------------------------------------")
    
    t0 = time.time()
    
    search.fit(X_train, y_train)
    
    best_estimator = search.best_estimator_.steps
    
    run_time = time.time() - t0
    print("------------------------------------------------------------------")
    print("Run time %0.3fs" % (run_time))
    
    print("Best score: %.3f" % search.best_score_)
    
    print("\n Parameters of the best model:")
    best_params = search.best_estimator_.get_params()
    for param_name in sorted(param_grid.keys()):
        print("\t%s: %r" % (param_name, best_params[param_name]))
    
    print("------------------------------------------------------------------")
    y_pred_train = search.predict(X_train)
    print("Classification report for the train set:")
    print(classification_report(y_train, y_pred_train))
     
    y_pred = search.predict(X_test)
    print("Classification report for the test set:")
    print(classification_report(y_test, y_pred))
    
    prec0, recall0, fscore0, _ = precision_recall_fscore_support(y_test, y_pred, 
                                                  pos_label=0, average='binary')
    
    prec1, recall1, fscore1, _ = precision_recall_fscore_support(y_test, y_pred, 
                                                  pos_label=1, average='binary')    

    print("------------------------------------------------------------------")
    
    roc_plot(search, X_test, y_test, classifier_name, n_classes=2)
    
    prec_recall_plot(search, X, y, classifier_name, n_classes=2)
    
    prec_recall_plot_multi(search, X, y, classifier_name, n_classes=2)
         
    return best_estimator
    
```

```python
def evaluation_blind_test(X_train, y_train, X_test, y_test, TfIdf_model, clf_model):
    
    tfvec = TfIdf_model
    tf_fit = tfvec.fit(X_train)
    X_train = tf_fit.transform(X_train).toarray()
    X_test = tf_fit.transform(X_test).toarray()
    
    clf_model.fit(X_train, y_train)
    
    y_pred = clf_model.predict(X_test)
    
    print("Classification report for the blind test set:")
    print(classification_report(y_test, y_pred))
    
    prec0, recall0, fscore0, support0 = precision_recall_fscore_support(y_test, y_pred, 
                                                  pos_label=0, average='binary')
    
    prec1, recall1, fscore1, support1 = precision_recall_fscore_support(y_test, y_pred, 
                                                  pos_label=1, average='binary')
    
    
    return prec0, recall0, fscore0, prec1, recall1, fscore1
```

```python
def evaluation_blind_test_resample(X_train, y_train, X_test, y_test, TfIdf_model, clf_model, resample_model):
    
    tfvec = TfIdf_model
    tf_fit = tfvec.fit(X_train)
    X_train = tf_fit.transform(X_train).toarray()
    X_test = tf_fit.transform(X_test).toarray()
    
    X_train_r, y_train_r = resample_model.fit_resample(X_train, y_train)
      
    clf_model.fit(X_train_r, y_train_r)
    
    y_pred = clf_model.predict(X_test)
    
    print("Classification report for the blind test set:")
    print(classification_report(y_test, y_pred))
    
    prec0, recall0, fscore0, support0 = precision_recall_fscore_support(y_test, y_pred, 
                                                  pos_label=0, average='binary')
    
    prec1, recall1, fscore1, support1 = precision_recall_fscore_support(y_test, y_pred, 
                                                  pos_label=1, average='binary')
    
    
    return prec0, recall0, fscore0, prec1, recall1, fscore1
```

```python
pipeline = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LogisticRegression(solver='liblinear', random_state=12))])

parameters = {
    'tfidf__max_df': (0.95, 0.9, 0.85, 0.8),
    'tfidf__min_df': (0.0001, 0.0005, 0.001, 0.002),
    'tfidf__norm': ('l1', 'l2'),
    'tfidf__use_idf': (True, False),
    'tfidf__smooth_idf': (True, False),
    'tfidf__sublinear_tf': (True, False),
#     'clf__penalty': ('l1', 'l2', 'elasticnet'),
    'clf__dual': (True, False),
    'clf__C': (1e-5, 1e2, "log-uniform"),
#     'clf_solver': ('newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'),
    'clf__max_iter': [100, 500, 1000, 2000, 10000],
    
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=12)

best_estimator_LR1 = bayessearchcv(
              pipeline=pipeline, param_grid=parameters, scoring='f1', cv=cv,
              n_iter=15, n_jobs=5, classifier_name='Logistic Regression')

print(best_estimator_LR1)
```

```python
TfIdf_model = best_estimator_LR1[0][1]

clf_model = best_estimator_LR1[1][1]

prec0_LR1, recall0_LR1, fscore0_LR1, prec1_LR1, recall1_LR1, fscore1_LR1 = evaluation_blind_test(
    X, y, X_test_blind, y_test_blind, TfIdf_model, clf_model)
```
